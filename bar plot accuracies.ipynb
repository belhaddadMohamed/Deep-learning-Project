```python
import matplotlib.pyplot as plt
import numpy as np
```

# 10 Models

```python
# Data
models = ['CNN-scratch', 'CNN-attention', 'VGG16', 'VGG19', 'InceptionV3', 'MobileNetV2', 'ResNet50', 'Xception', 'ViT-DeiT', 'DenseNet121']
nct_crc = [92.9, 10, 99, 98, 98.6, 98.1, 99.2, 98, 97, 96]
crc_val = [80, 10, 94, 94, 95, 94, 94, 94, 91, 96]

# Set width of bars
barWidth = 0.4

# Set position of bar on X axis
r1 = np.arange(len(nct_crc))
r2 = [x + barWidth for x in r1]

# Create plot
fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(r1, nct_crc, color=['#738ADB', '#738ADB', '#738ADB', '#738ADB', '#738ADB', '#738ADB', '#738ADB', '#738ADB', '#738ADB', '#96D7B4'], width=barWidth, edgecolor='white', label='NCT-CRC-HE-100K')
rects2 = ax.bar(r2, crc_val, color=['#A5C8E1', '#A5C8E1', '#A5C8E1', '#A5C8E1', '#A5C8E1', '#A5C8E1', '#A5C8E1', '#A5C8E1', '#A5C8E1', '#96D7B4'], width=barWidth, edgecolor='white', label='CRC-VAL-HE-7K')

# Add xticks on the middle of the group bars
plt.xlabel('Model', fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(nct_crc))], models)

# Add y-axis labels and title
plt.ylabel('Accuracy Rate (%)', fontweight='bold')
plt.title('Comparison of Model Performance on NCT-CRC-HE-100K and CRC-VAL-HE-7K Datasets', fontweight='bold')

# Create legend and show plot
plt.legend()
plt.show()
```
